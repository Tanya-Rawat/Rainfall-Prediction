## EDA for weather.csv

### 1. Importing and Loading the Dataset

```python
import pandas as pd
data = pd.read_csv("data/weather.csv")
data.head()
```

### 2. Summary Statistics

```python
data.describe()
```

### 3. Data Types and Missing Values

```python
data.shape
data.info()
data.isnull().sum()
```

```python
data.duplicated().sum()
```

### 4. Identifying numerical and categorical features


```python
# Categorical feature
category_feature = [feature for feature in data.columns if data[feature].dtypes == 'O']
category_feature
```

```python
# Numerical feature
numerical_feature = [feature for feature in data.columns if data[feature].dtypes != 'O']
numerical_feature
```

### 5. Distribution of Numerical Features

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograms for numerical features
data.hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

### 6. Correlation Analysis of Numerical Features

```python
# Remove rows with 'object' data type
df_numeric = data.select_dtypes(exclude=['object'])

# Calculate the correlation matrix
corr_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
sns.clustermap(corr_matrix)
```

### 7. Distribution of Categorical Features

```python
# Count plot for categorical features
plt.figure(figsize=(10, 6))
sns.countplot(x='RainToday', data=data)
plt.title("RainToday Distribution")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='RainTomorrow', data=data)
plt.title("RainTomorrow Distribution")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='WindGustDir', data=data)
plt.title("WindGustDir Distribution")
plt.xticks(rotation=45)
plt.show()


```

### 8. Outlier Detection

```python
# Box plots for numerical features
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.show()
```
